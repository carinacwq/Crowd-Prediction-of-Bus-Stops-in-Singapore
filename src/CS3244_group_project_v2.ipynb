# Crowd Prediction of Bus Stops in Singapore
Public transportation systems, such as buses, are integral to urban mobility, particularly in densely populated countries like Singapore. Understanding and predicting the usage trends and patterns at bus stops can aid in the optimisation of bus schedules, improve service reliability and enhance the overall commuter experience.

This project aims to create a machine learning-based system that accurately predicts the expected passenger crowd levels at a bus stop at a particular time in Singapore, enabling authorities and planners to make informed data-driven decisions on bus stop crowd control. This includes leveraging historical data of different bus stops, including information such as tap-in-tap-out passenger volume, day of the week, public holidays.

## Data Pre Processing
To prepare for further analysis, the Passenger Volume dataset was joined with the Bus Stop Locations dataset on the `BusStopCode`. To clean the data, we also removed observations with null values and converted the data types for relevant columns like `TIME_PER_HOUR`. Moreover, a new variable `avg_volume` was created from Tap-In and Tap-Out Volume as the Average Passenger Volume will serve as the target variable to be predicted. \
Our final dataset consists of about 600,000 observations with features including passenger volume, day of the week (weekday or weekend/public holiday), time of the day (hour), bus stop locations and planning areas. 

```python
#load data locally
pv_df = pd.read_csv("../data/pv.csv")
busstops_df = pd.read_csv("../data/busstops_with_planningarea.csv")
print(pv_df.head(5))
print(pv_df.info())
print(busstops_df.head(5))
print(busstops_df.info())

# Join both df
pv_df['PT_CODE'] = pv_df['PT_CODE'].apply(lambda x: '{:05d}'.format(x))
busstops_df['BusStopCode'] = busstops_df['BusStopCode'].apply(lambda x: '{:05d}'.format(x))
merged_df = pv_df.merge(busstops_df, left_on='PT_CODE', right_on='BusStopCode', how='left')
print(merged_df.head(10))
print(merged_df.info())
```

### Remove Null Values
Upon research, `PT_CODE` = 65139 has been removed since Dec '23 due to route amendment

```python
# Dealing with Null
null_df=merged_df[merged_df.isnull().any(axis=1)]
# print(null_df)

# Remove rows with PT_CODE 65139 and assign results back to original df
merged_df = merged_df.drop(merged_df[merged_df['PT_CODE'] == '65139'].index)
# Drop rest of null values
merged_df = merged_df.dropna()
print(merged_df.info())
```

### Further Cleaning

```python
# Convert TIME_PER_HOUR to int
merged_df['TIME_PER_HOUR'] = merged_df['TIME_PER_HOUR'].astype(int)
# Drop duplicate column
merged_df = merged_df.drop(columns=['BusStopCode'])

# Create a new column for average passenger volume
merged_df['avg_volume'] = (merged_df['TOTAL_TAP_IN_VOLUME'] + merged_df['TOTAL_TAP_OUT_VOLUME']) / 2
merged_df.head()
```

## Feature Engineering
Since we are performing classification to predict the crowd levels at the bus stops, we binned the values in `avg_volume` into 2 labels: 0 (Low), 1 (High). \
Since most of the variables in our dataset are categorical, we performed one-hot encoding on features like `planning_area` and `Day_Type` to transform the features into a numerical format to be fed into the ML models. \
We also performed a train-test split of the data in a 80/20 ratio for the model development. 

### Bin `avg_volume` into labels for classification
0 - Low, 1 - High

```python
from sklearn.preprocessing import LabelEncoder

y = merged_df['avg_volume']

# Define bin boundaries
num_bins = 3
bin_boundaries = np.percentile(y, np.linspace(0, 100, num_bins))

# Assign labels
labels = np.digitize(y, bin_boundaries)

# Encode labels
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Add last value to Label = 1
encoded_labels[np.where(encoded_labels == 2)] = 1
```

### One-Hot Encoding to encode categorical features - Planning Area, Day Type

```python
# Define features and target variable
X = merged_df[['DAY_TYPE', 'TIME_PER_HOUR', 'planning_area']]

# Perform one-hot encoding for categorical features
cat_features = ['DAY_TYPE', 'planning_area']
X_encoded = pd.get_dummies(X, columns=cat_features, drop_first=True, dtype=float)
X_encoded.info()
X_encoded.head()
```

### Split train and test sets

```python
from sklearn.model_selection import train_test_split
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, encoded_labels, test_size=0.2, random_state=42)
```

## Model Development

As part of our model development process, we used Logistic Regression, k-Nearest Neighbor (kNN) Classification and Random Forest Classification to train our models.


### Logistic Regression
Logistic Regression was run as our baseline model, assuming linear relationship between Crowd Levels and the independent features.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create a logistic regression model
log_reg = LogisticRegression(max_iter=1000)

# Train the model
log_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = log_reg.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

#### Receiver Operator Characteristic (ROC) Curve 

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities for the test set
y_probs = log_reg.predict_proba(X_test)
# Keep probabilities for the positive class only
y_probs = y_probs[:, 1]

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calculate the AUC (Area Under Curve)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC Curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression')
plt.legend(loc='lower right')
plt.show()
```

### k-Nearest Neighbor (kNN) Classification

For k-Nearest Neighbors (KNN) Classification, we first scaled the training and testing data, chose the value of k using the rule of thumb before conducting hyperparameter tuning via the grid search method. We then fitted the kNN classifier to the training data and evaluated the modelâ€™s performance.

#### Standardise values in `X_train`, `X_test` features

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the training data and transform X_train
X_train_scaled = scaler.fit_transform(X_train)

# Transform X_test using the same scaler
X_test_scaled = scaler.transform(X_test)
```

#### Hyperparameter Tuning

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

k = int(np.sqrt(len(X_train))) # Rule of thumb for k

# Define hyperparameter grid
param_grid = {
    'n_neighbors': [5, 7, 9, int(k/2), k]
}

# Create KNN regressor
knn_clf = KNeighborsClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(knn_clf, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Get best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

#### Final Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=345)  

# Train the classifier
knn_classifier.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = knn_classifier.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

#### Receiver Operator Characteristic (ROC) Curve 

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities for the test set
y_probs = knn_classifier.predict_proba(X_test_scaled)
# Keep probabilities for the positive class only
y_probs = y_probs[:, 1]

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calculate the AUC (Area Under Curve)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC Curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for k-Nearest Neighbours(kNN)')
plt.legend(loc='lower right')
plt.show()
```

### Random Forest Classification

Random Forest Classification, consisting of an ensemble of decision trees was also used in attempts to better capture the non-linear relationships present between our features and target variables. Hyperparameter tuning was performed using GridSearchCV with cross-validation to find the best parameters. The final model was built with the best paramters found during hyperparameter tuning. Predictions were made using the test data and we evaluated the model's performance using accuracy. Feature importances are also analysed to determine the most important features in the model.

#### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# define parameter grid to search
param_grid_rf = {
    'n_estimators': [10, 20, 30],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
} 


# create Random Forest Regressor model
rf_clf = RandomForestClassifier(random_state = 42)

# instantiate GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf_clf, param_grid=param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)

# perform Grid Search to find best parameters for hyperparameter tuning
grid_search_rf.fit(X_train, y_train)

# find best parameters
print("Best Parameters:", grid_search_rf.best_params_)
```

#### Final Model

```python
import joblib
from sklearn.ensemble import RandomForestClassifier

# build model with best paramters
best_rf = RandomForestClassifier(n_estimators=30, max_depth=None, min_samples_split=10, random_state=42)

# fit model to training data
best_rf.fit(X_train, y_train)

# save model
joblib.dump(best_rf, 'models/best_rf_model.pkl')
```

#### Evaluation

```python
import joblib
# Load the model
best_rf = joblib.load('models/best_rf_model.pkl')

# Evaluate the model
accuracy = best_rf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

#### Receiver Operator Characteristic (ROC) Curve 

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities for the test set
y_probs = best_rf.predict_proba(X_test)
# Keep probabilities for the positive class only
y_probs = y_probs[:, 1]

# Calculate the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calculate the AUC (Area Under Curve)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Random Forest')
plt.legend(loc='lower right')
plt.show()
```

## Discussion and Conclusion

**Analysis of results using Accuracy**

Among the 3 models, the Random Forest model seems to perform the best in predicting passenger crowd levels at bus stops given its highest accuracy of about 0.732, as compared to the accuracy of the Logistic Regression model at about 0.669 and that of kNN at about 0.729. This might be attributed to the fact that it combines the predictions of multiple decision trees, which helps mitigate overfitting and reduces the variance of the model, leading to more accurate predictions. The kNN model also performs relatively well and this could be due to the fact that the data is well-structured. In this case where bus stop crowd levels are influenced by similar factors across different instances, kNN's ability to classify based on proximity in feature space might be effective.

**Analysis of results using ROC Curve**

Among the 3 models, the kNN Classifier and Random Forest model have the higher value of Area Under Curve (AUC) at 0.81, compared to that of the Logistic Regression model at 0.74. This shows that the kNN Classifier and Random Forest model perform similarly, but both perform better than the Logistic Regression model in predicting passenger crowd levels at bus stops. This could be because kNN and Random Forest are both non-linear models that can capture complex relationships in the data without assuming linearity, and in the context of predicting crowd levels at bus stops, these models can better handle non-linear relationships between various factors and crowd levels.

**Pros and Cons of Models**

Logistic Regression offers simplicity in its implementation and rapid training even with large datasets, making it an appealing choice as our baseline model for classification tasks. It provides a straightforward approach to analyze the relationship between multiple predictor variables such as TIME_PER_HOUR, DAY_TYPE, etc., and our predicted categorical variable for average passenger volume. However, its simplicity can also be a limitation, as it may fail to capture the intricacies and nonlinearities present in real-world data. Logistic Regression is particularly sensitive to outliers, which is a highly probable occurrence in the case of average passenger volume data due to one-time off events happening around a planning area.

The kNN algorithm is easy to understand and implement in such classification tasks, as it does not require any assumptions about the underlying data distribution of crowd levels. It is also robust to noisy data and can effectively handle them because it doesn't rely on assumptions about the distribution of the data. However, the choice of the parameter k (the number of neighbors) significantly impacts the performance of the kNN algorithm, thus selecting an appropriate value for k is crucial. In addition, since kNN makes predictions by calculating distances to all training points, it can be computationally expensive, especially with high-dimensional feature spaces or large datasets like in this case, where we are working with 600,000 unique observations together with a number of features including passenger volume, day of the week, time of the day, bus stop locations and planning areas.

The random forest classifier is less prone to overfitting and with the decision trees built incrementally, it will be able to capture non-linear relationships more effectively than linear models like logistic regression. By aggregating predictions from multiple decision trees, random forest classifier exhibits resilience to noisy data. Due to the ensemble nature of Random Forest, individual noisy data points are less likely to significantly impact the predictions of every tree. As a result, the overall performance of the model is less susceptible to the influence of noisy data. However, due to its complexity and the large number of decision trees, it can be computationally expensive to train the Random Forest Classifier on a large dataset. Random Forest models may also require more time to generate predictions compared to other algorithms, despite being more effective at training. This is especially salient when dealing with large datasets or models consisting of numerous trees. In order to produce a final prediction, each data point must traverse through multiple decision trees within the forest. Consequently, this process may prolong the prediction time, particularly in scenarios where real-time or latency-sensitive applications are involved.

**Conclusion**

From the results above, both kNN and Random Forest models demonstrate relatively high accuracy and AUC values, outperforming Logistic Regression. The kNN and Random Forest models also show comparable performance in predicting passenger volume at bus stops, with both achieving AUC values of 0.81. These results suggest that both kNN and Random Forest models are effective for the classification task of predicting passenger volume at bus stops.

However, the choice between kNN and Random Forest may depend on other considerations such as computational complexity, interpretability, and ease of implementation. Collecting other features such as weather conditions and population density, and further analysis including examining additional evaluation metrics and considering practical implications, may be necessary to make a final decision on the most suitable model for deployment in predicting passenger volume at bus stops.
